<h1 align=center> Ensemble Learning In Depth </h1>



- **Ensemble Learning** involves combining multiple models to improve predictive performance
- Ensemble learning can be use in classification and regression problems
- Ensemble models are effective in handling imbalanced datasets
- Ensembles improve model robustness and stability by reducing the impact of outliers and noise in the data
- Reduce overfitting
- Ensemble models often generalize better to unseen data compared to single models

### **Types of Ensemble Learning:**



1. **Bagging (Bootstrap Aggregating)**:
    - In bagging, multiple base learners are trained independently on different subsets of the training data, which are sampled with replacement (bootstrap samples)
    - It is parallel ensemble learning
    - The final prediction is obtained by averaging (for regression) or voting (for classification) over the predictions of all base learners
    - Decrease the variance
    - Example algorithms: Random Forest



1. **Boosting**:
    - Boosting sequentially trains a series of weak learners, with each subsequent learner focusing more on the training instances that were misclassified by the previous learners
    - Or each new model tries to learn from the errors of the previous models in the ensemble
    - Sequential ensemble learning
    - The goal is to improve the overall performance by focusing on areas where previous models made mistakes
    - The final prediction is typically a weighted sum of the predictions of all weak learners
    - Reduce the bias
    - Example algorithms: AdaBoost, Gradient Boosting Machines (GBM), XGBoost, LightGBM, CatBoost.





1. **Stacking (Stacked Generalization)**:
    - Stacking combines the predictions of multiple base learners using a meta-learner (also called a blender or a combiner).
    - The base learners are trained on the original data, and their predictions are then used as features to train the meta-learner
    - The meta-model then makes the final predictions on new data
    - Example algorithms: Stacked Ensemble Models.
