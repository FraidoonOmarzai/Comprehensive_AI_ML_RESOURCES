### 1. Mean Absolute Error (MAE)

- Calculate the absolute difference between actual and predicted values

**Advantages:**

- Same unit
- Robust to outlier

**Disadvantages:**

- Fails to punish the bigger error terms
- The graph is not differentiable, so we have to apply various optimizers like gradient decent

$$
MAE = \frac{1}{n} \sum_{i=1}^n | y_i - \hat{y_i}|
$$

- y_hat: predicted value
- y_i: actual value for the ith data point
- n: number of observations

### 2. Mean Square Error (MSE)

- Finding the squared difference between actual and predicted values
- The MSE punishes larger errors more than smaller errors

**Advantages:**

- Graph is differentiable

**Disadvantages:**

- Not robust to outlier because it penalizes the outlier too much
- Value getting after the calculation is a squared unit of output

$$
MSE = \frac{1}{n} \sum_{i=1}^n ( y_i - \hat{y_i})^2
$$

- y_hat: predicted value
- y_i: actual value for the ith data point

### 3. Root Mean Square Error (RMSE)

- RMSE is the root square of MSE
- The RMSE punishes larger errors more than smaller errors

**Advantages:**

- Same unit

**Disadvantages:**

- Not that robust to outliers as compared to MSE

$$
RMSE =\sqrt{\frac{1}{n} \sum_{i=1}^n ( y_i - \hat{y_i})^2}
$$

### 4. R-Squared (R2)

- R-squared shows how well the data fit the regression model (the goodness of fit)
- 0 =< R2 =< 1
- The drawback is when we add new features to data the R2 score starts increasing or remains constant but it never decreases, because it assumes that while adding more data variance of data increase

$$
R^2 = 1 - \frac{SS_{Regression}}{SS_{Total}} = 1 - \frac{\sum_{i=1}^n ( y_i - \hat{y_i})^2}{\sum_{i=1}^n ( y_i - \overline{y})^2}
$$

- y_hat: predicted value
- y_dash:  mean of actual value

### 5. Adjusted R-Squared

- The adjusted R-squared shows whether adding additional predictors improves a regression model or not

$$
Adjusted\; R^2 = 1- \frac{(1-R^2)(N-1)}{N-p-1}
$$

- R^2: sample R-squared
- N: total sample size
- p: number of independent variable
