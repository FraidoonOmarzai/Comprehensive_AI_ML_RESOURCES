<h1 align=center> Logistic Regression In Depth </h1>



- Supervised ML algorithm
- Used for binary classification
- Great starter algorithm for text related classification
- Improve by applying feature scaling
- Not much effected by outliers due to the sigmoid function tapers the outliers
- The output value is between 0 and 1

#### 1. **Train the algorithm**


$$
z = wx + b
$$

$$
g(z) = \hat y = f_{w,b}(x) = \frac 1{1+e^{-z}}
$$

- *g(z):* is the predicted value
- *w:* slope or coefficient
- *x:* independent variables
- *b:* constant or intercept

### 2. **Measure the accuracy of the model**

`Note`: In logistic regression, we use different loss or cost function compare to linear regression due to having multiple local minimum (non-convex function). 

- Loss function: measure error for single training example

$$
L(f_{w,b}(x^{(i)}), y^{(i)})=-y^{(i)} \log {f_{w,b}(x^{(i)})}-(1-y^{(i)}) \log (1-{f_{w,b}(x^{(i)})})
$$

- cost function: is the average of the loss function of the entire training set

$$
J(w,b)=- \frac1{m} \sum \limits _{i=1} ^{m} [y^{(i)} \log {f_{w,b}(x^{(i)})}+(1-y^{(i)}) \log (1-{f_{w,b}(x^{(i)})})]
$$

### 3. **Minimize the error**
- We use gradient descent algorithms to minimize cost function or the error between predicted and actual values in various machine learning algorithms
- Gradient descent is an optimization algorithm
- It updates the values of w and b

`Note:` We need to perform feature scaling when dealing with gradient descent-based algorithms (linear regression, logistic regression, neural network) and distance-based algorithms (KNN, K-means, SVM) as they are sensitive to the range of the data points.




$$
w = w- \alpha \frac d{dw} J(w,b)
$$

$$
b = b- \alpha \frac d{db} J(w,b)
$$

- In the above equation, alpha determines the step size at each iteration while moving toward a minimum loss function and it is a tuning parameter in an optimization algorithm.

### Pros 
- Simplicity and Interpretability
- Efficiency
- Regularization Techniques: Logistic regression can incorporate regularization techniques (such as L1 and L2 regularization) to prevent overfitting and handle multicollinearity.
- Handling Multiclass Problems: Though inherently a binary classifier, logistic regression can be extended to multiclass classification problems using techniques like One-vs-Rest (OvR) or One-vs-One (OvO).

Cons
- Assumption of Linearity: Logistic regression assumes a linear relationship between the independent variables and the log odds. It may not perform well if the true relationship is not linear.
- Sensitivity to Outliers
- Need for Feature Scaling
- Overfitting with High Dimensionality

```python
# import the iris dataset and logistic regression from sklearn 
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression

# load the dataset
X, y = load_iris(return_X_y=True)

# train the model
clf = LogisticRegression(random_state=0).fit(X, y)

# measure the accuracy
clf.score(X, y)
```
