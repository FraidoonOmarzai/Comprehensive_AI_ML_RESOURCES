<h1 align=center> Common Optimization Algorithms </h1>



 Optimization algorithms are crucial for training deep learning models by updating weights and biases to minimize the loss function.

- Epochs: number of iteration.
- Batch size: number of training examples utilized in one iteration.

### 1. Stochastic Gradient Descent(SGD)

- When the batch size is the size of one sample.
- More prone to getting stuck in local minima.

### 2. Mini-Batch GradientDescent

- The batch size is more than one sample and less than the size of the training dataset.
- Faster learning.

### 3. Batch GradientDescent

- When you process your entire training set all at the same time.
- Too long per iteration.



### 4. Momentum

- This method helps accelerate gradient vectors in the right directions, thus leading to faster converging.
- It improves the performance of gradient descent.



### 5. RMSProp (Root Mean Square Propagation)

- It is an adaptive optimization algorithm.
- Well suited for problems with noisy gradients.



### 6. Adam (Adaptive Moment Estimation)

- Combines the ideas of momentum and RMSprop.
- Adaptive Optimization algorithm.
- Widely used.
- Effective, but can be computationally expensive.
