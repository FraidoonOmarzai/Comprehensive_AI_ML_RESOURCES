# Skewness

- The skewness is a measure of the asymmetry of the distribution

- In statistic, symmetry refers to the balance or equality of value on both sides of a central point in a distribution.

![alt text](../Images/math/statistic/sk.png)

1. **Right Skewness**
    1. Positive skew
    2. The tail is longer on the right
    3. mean>median>mode
2. **Left Skewness**
    1. Negative skew
    2. The tail is longer on the left
    3. mode>median>mean
3. **Symmetric Distribution**
    1. Mean, median and mode are all equal 
    2. The left and right tails of the distribution are equal in length and shape 
    3. The data points are evenly distributed on both sides of the central point
    4. mean=median=mode

### Covariance

- One of the very important topic of data preprocessing, and it is used to evaluate the linear relationships between two variables

![alt text](../Images/math/statistic/Covar.png)

### **Pearson Correlation Coefficient (Pearson Covariance)**

- It is a standardized measure of the linear relationship between two variables.
- Pearson correlation is calculated as the covariance of the variables divided by the product of their standard deviations
- It is sensitive to outliers and assumes that the relationship between variables is linear
- Result always has a value between −1 and 1

![alt text](../Images/math/statistic/p_corv.webp)

### **Spearman Rank Correlation Coefficient (Spearman's Rank)**

- Assesses monotonic relationships using ranks rather than the actual values
- Spearman's rank correlation is particularly useful when the variables do not follow a normal distribution or when there are outliers present

### Hypothesis Testing

![alt text](../Images/math/statistic/hypothesis.png)

- It is a way to tell if some belief you have about the population is likely to be true or false
- Hypothesis testing is a tool for making statistical inferences about the population data
- Hypothesis testing is the evaluation of two mutually exclusive statements or populations using sample data
- Null Hypothesis (H0): default assumption or baseline, treats everything the same or equal
- Alternative Hypothesis (H1): what you want to prove

### Type I and Type II Errors

![alt text](../Images/math/statistic/type12.png)

- **Errors:** There's a chance of making errors:
    - **Type I error (α):** Rejecting a true null hypothesis (false positive)
    - **Type II error (β):** Failing to reject a false null hypothesis (false negative)

`Note`: What is the greatest probability of type I error you are willing to tolerate? ⇒ Significance value (**α**)

### **Right-Tailed, Left-Tailed, and Two-Tailed Tests**

![alt text](../Images/math/statistic/righttest.png)

### **1. Right-Tailed Test (Upper-Tailed Test):**

- **Directionality**: The critical region is on the right side of the distribution.
- **Null Hypothesis (H0)**: Typically states a condition of "no effect" or "no difference."
- **Alternative Hypothesis (H1)**: States that the parameter of interest is greater than a specified value.
- **Decision Rule**: Reject the null hypothesis if the test statistic falls in the right tail of the distribution.
- **Example**: Testing whether a new drug treatment leads to an increase in patient recovery rates.

### **2. Left-Tailed Test (Lower-Tailed Test):**

- **Directionality**: The critical region is on the left side of the distribution.
- **Null Hypothesis (H0)**: Typically states a condition of "no effect" or "no difference."
- **Alternative Hypothesis (H1)**: States that the parameter of interest is less than a specified value.
- **Decision Rule**: Reject the null hypothesis if the test statistic falls in the left tail of the distribution.
- **Example**: Testing whether a new teaching method results in a decrease in student test scores.

### **3. Two-Tailed Test:**

- **Directionality**: The critical region is split between the left and right sides of the distribution.
- **Null Hypothesis (H0)**: Typically states a condition of "no effect" or "no difference."
- **Alternative Hypothesis (H1)**: States that the parameter of interest is different from a specified value (either greater or less than).
- **Decision Rule**: Reject the null hypothesis if the test statistic falls in either tail of the distribution.
- **Example**: Testing whether a coin is biased (two-tailed) by flipping it and checking for deviations from an expected 50-50 heads-tails distribution.


### **Steps in Hypothesis Testing:**

1. **Formulate Hypotheses**: State the null hypothesis and alternative hypothesis
2. **Select a Significance Level (α)**: Choose a threshold for rejecting the null hypothesis, typically 0.05
    - **P-value:** This is the probability of observing your data (or more extreme) if the null hypothesis were true. A low p-value (typically below 0.05) suggests rejecting the null hypothesis
3. **Collect Data**: Gather sample data relevant to the hypothesis being tested
4. **Calculate Test Statistic**: Compute a test statistic that measures the likelihood of obtaining the observed data under the null hypothesis
5. **Make a Decision**: Compare the test statistic to a critical value or calculate a p-value. If the test statistic falls in the rejection region (i.e., beyond the critical value) or if the p-value is less than the significance level, reject the null hypothesis in favor of the alternative hypothesis. Otherwise, fail to reject the null hypothesis
6. **Draw Conclusion**: Based on the decision, draw conclusions about the hypothesis being tested

### **Example:**

Suppose you have a coin, and you want to test whether it's a fair coin (i.e., has an equal probability of landing heads or tails). You decide to conduct an experiment where you flip the coin 100 times and record the outcomes

**Hypotheses:**

- Null Hypothesis (H0): The coin is fair (p = 0.5, where p is the probability of heads)
- Alternative Hypothesis (H1): The coin is not fair (p ≠ 0.5)

**Experiment:**

You flip the coin 100 times and observe that it lands heads 57 times and tails 43 times.

**Test Statistic:**

You can use the binomial distribution to calculate the probability of getting 57 or more heads out of 100 flips if the coin is fair (under the null hypothesis)

**Decision Rules:**

Let's choose a significance level (α) of 0.05. If the probability of getting 57 or more heads is less than 0.05, we'll reject the null hypothesis in favor of the alternative hypothesis

**Calculation:**

Using the binomial distribution, the probability of getting 57 or more heads out of 100 flips with a fair coin is approximately 0.024

`Note`: To find the probability of getting 57 or more heads out of 100 flips with a fair coin (under the null hypothesis), we need to calculate the cumulative probability from 57 to 100 using the binomial distribution formula:

*P*(*X*≥57)=*P*(*X*=57)+*P*(*X*=58)+...+*P*(*X*=100)

We'll sum up the individual probabilities for *k* ranging from 57 to 100. Each individual probability is calculated using the binomial probability formula:

![alt text](<../Images/math/probability/binomial formula.png>)

- *n*=100 (number of coin flips)
- *p*=0.5 (probability of heads)

**Conclusion:**

Since the calculated probability (0.024) is less than the significance level (0.05), you reject the null hypothesis. Therefore, you conclude that there is evidence to suggest that the coin is not fair, and it may be biased towards landing heads

`Note`: In this example, hypothesis testing helps you make an inference about the fairness of the coin based on observed data, using probability theory to quantify the likelihood of the observed outcome under different hypotheses.

### Statistical Testing

1. **T-test:**
    - Comparison of mean of two groups, where sample SD is given
    - It is similar to the Z-statistic but uses the sample standard deviation
    - Note that you can only uses a t test to compare two means. If you want to compare three or more means, use an ANOVA-test instead
    - If the sample size is less than 30, we use t-test
    
    **There are several types of t-tests:**
    
    1. **Independent Samples t-test**: This test compares the means of two independent groups to determine if they are significantly different from each other. For example, you might use an independent samples t-test to compare the test scores of two different classes to see if there's a significant difference in performance.
    2. **Paired Samples t-test**: Also known as a dependent samples t-test, this test compares the means of two related groups to determine if there is a significant difference between them. For example, you might use a paired samples t-test to compare the before and after scores of students who underwent a tutoring program.
    3. **One-Sample t-test**: This test compares the mean of a single sample to a known value or a hypothetical value. For example, you might use a one-sample t-test to determine if the average weight of a sample of apples differs significantly from 150 grams, the standard weight of an apple.
