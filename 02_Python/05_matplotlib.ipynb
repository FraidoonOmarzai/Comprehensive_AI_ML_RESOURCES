### 1. Line Plot

- **Use Case**: To show trends over time or a continuous variable.
- **Example**: Plotting stock prices over time.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("Time (Days)")
plt.ylabel("Stock Price ($)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
ax.plot(x, y1, label='Sine Wave', color='b', linestyle='-', linewidth=2, marker='o')
ax.plot(x, y2, label='Cosine Wave', color='r', linestyle='--', linewidth=2, marker='x')

# Add title and labels
ax.set_title('Sine and Cosine Waves', fontsize=16)
ax.set_xlabel('X-axis', fontsize=14)
ax.set_ylabel('Y-axis', fontsize=14)

# Customize the grid
ax.grid(True, linestyle='--', alpha=0.6)

# Add legend
ax.legend(loc='upper right', fontsize=12)

# Annotate a point
ax.annotate('Max of Cosine', xy=(0, 1), xytext=(2, 1.2),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Show the plot
plt.show()
```

**Subplot**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

# Create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot 1: Sine Wave
axs[0, 0].plot(x, y1, 'b', label='Sine')
axs[0, 0].set_title('Sine Wave')
axs[0, 0].legend()

# Plot 2: Cosine Wave
axs[0, 1].plot(x, y2, 'r--', label='Cosine')
axs[0, 1].set_title('Cosine Wave')
axs[0, 1].legend()

# Plot 3: Tangent Wave (with limits to avoid extreme values)
axs[1, 0].plot(x, y3, 'g-.', label='Tangent')
axs[1, 0].set_ylim(-10, 10)
axs[1, 0].set_title('Tangent Wave')
axs[1, 0].legend()

# Plot 4: Exponential Decay
axs[1, 1].plot(x, y4, 'm-', label='Exponential Decay')
axs[1, 1].set_title('Exponential Decay')
axs[1, 1].legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Show plot
plt.show()

```

### 2. Scatter Plot

- **Use Case**: To show relationships or correlations between two variables.
- **Example**: Examining the relationship between height and weight of individuals.

```python
import matplotlib.pyplot as plt

height = [150, 160, 170, 180, 190]
weight = [50, 60, 65, 70, 80]

plt.scatter(height, weight)
plt.title("Scatter Plot Example")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()
```

### 3. Bar Plot

- **Use Case**: To compare quantities corresponding to different groups.
- **Example**: Comparing the number of sales of different products.

```python
import matplotlib.pyplot as plt

products = ['A', 'B', 'C', 'D']
sales = [30, 40, 35, 50]

plt.bar(products, sales)
plt.title("Bar Plot Example")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

### 4. Histogram

- **Use Case**: To show the distribution of a dataset.
- **Example**: Plotting the distribution of exam scores of students.

```python
import matplotlib.pyplot as plt
import numpy as np

scores = np.random.normal(70, 10, 100)

plt.hist(scores, bins=10)
plt.title("Histogram Example")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Generate random data
data = np.random.normal(loc=0, scale=1, size=1000)

# Create a figure
plt.figure(figsize=(10, 6))

# Plot histogram using matplotlib
plt.hist(data, bins=30, color='skyblue', alpha=0.7, edgecolor='black', density=True, label='Histogram')

# Plot density plot using seaborn
sns.kdeplot(data, color='red', linewidth=2, label='Density Plot')

# Add title and labels
plt.title('Histogram and Density Plot', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Density', fontsize=14)

# Add legend
plt.legend()

# Show plot
plt.show()

```

### 5. Box Plot

- **Use Case**: To show the spread and skewness in data, as well as identify outliers.
- **Example**: Analyzing the distribution of salaries in a company.

```python
import matplotlib.pyplot as plt

salaries = [40000, 45000, 50000, 55000, 60000, 70000, 80000, 90000, 120000]

plt.boxplot(salaries)
plt.title("Box Plot Example")
plt.ylabel("Salary ($)")
plt.show()
```

### 6. Pie Chart

- **Use Case**: To show the proportions of categorical data.
- **Example**: Displaying the market share of different companies.

```python
import matplotlib.pyplot as plt

market_share = [35, 30, 20, 15]
companies = ['Company A', 'Company B', 'Company C', 'Company D']

plt.pie(market_share, labels=companies, autopct='%1.1f%%')
plt.title("Pie Chart Example")
plt.show()
```

### 7. Heatmap

- **Use Case**: To show correlations between different variables or to display data in a matrix form.
- **Example**: Visualizing a correlation matrix in a dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = np.random.rand(10, 10)
sns.heatmap(data, annot=True, cmap='viridis')
plt.title("Heatmap Example")
plt.show()
```

### 8. Violin Plot

- **Use Case**: To show the distribution of data across different categories, combining aspects of box plots and histograms.
- **Example**: Analyzing the distribution of petal lengths of different species of flowers.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.violinplot(data=data)
plt.title("Violin Plot Example")
plt.show()
```

### 9. Area Plot

- **Use Case**: To show cumulative data over time.
- **Example**: Displaying the cumulative sales over a year.

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [100, 150, 200, 250, 300, 350]

plt.fill_between(months, sales, color='skyblue', alpha=0.4)
plt.plot(months, sales, color='Slateblue', alpha=0.6)
plt.title("Area Plot Example")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

### 10. Stem Plot

- **Use Case**: To visualize discrete data points.
- **Example**: Displaying the rainfall on different days.

```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5]
rainfall = [0, 20, 0, 10, 30]

plt.stem(days, rainfall)
plt.title("Stem Plot Example")
plt.xlabel("Days")
plt.ylabel("Rainfall (mm)")
plt.show()
```

### 11. Step Plot

- **Use Case**: To show how values change at specific points.
- **Example**: Visualizing how stock prices change with time intervals.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.step(x, y)
plt.title("Step Plot Example")
plt.xlabel("Time (Days)")
plt.ylabel("Stock Price ($)")
plt.show()
```

### 12. 3D Surface Plot

- **Use Case**: To visualize three-dimensional data.
- **Example**: Showing the peaks and valleys of a mathematical function.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
plt.title("3D Surface Plot Example")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Generate random data
np.random.seed(0)
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)
color = np.random.rand(100)

# Create a figure
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot a 3D scatter plot
sc = ax.scatter(x, y, z, c=color, cmap='viridis', s=50)

# Add color bar
plt.colorbar(sc)

# Set labels
ax.set_title('3D Scatter Plot', fontsize=16)
ax.set_xlabel('X-axis', fontsize=12)
ax.set_ylabel('Y-axis', fontsize=12)
ax.set_zlabel('Z-axis', fontsize=12)

# Show plot
plt.show()

```
