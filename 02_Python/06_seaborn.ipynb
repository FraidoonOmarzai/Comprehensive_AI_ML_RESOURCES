### Common Plots in Seaborn

Seaborn offers several types of plots that are commonly used for data visualization:

1. **Distribution Plots**: `histplot`, `kdeplot`, `displot`
2. **Categorical Plots**: `barplot`, `countplot`, `boxplot`, `violinplot`, `stripplot`, `swarmplot`
3. **Relational Plots**: `scatterplot`, `lineplot`
4. **Regression Plots**: `regplot`, `lmplot`
5. **Matrix Plots**: `heatmap`
6. **Pairwise Plots**: `pairplot`
7. **Facet Grids**: `FacetGrid`

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
tips = sns.load_dataset('tips')
tips.head()
```

This dataset contains the following columns:

- `total_bill`: Total bill in dollars
- `tip`: Tip given in dollars
- `sex`: Gender of the person
- `smoker`: Whether the person is a smoker or not
- `day`: Day of the week
- `time`: Time of day (Lunch or Dinner)
- `size`: Number of people at the table

### 1. Distribution Plots

**Histogram (`histplot`)**

A histogram shows the distribution of a dataset. It groups the data into bins and shows the number of observations in each bin.

```python
sns.histplot(tips['total_bill'], kde=True)
plt.title('Distribution of Total Bill')
plt.show()
```

- `kde=True`: Adds a Kernel Density Estimate (KDE) line over the histogram to show the distribution's shape more smoothly.

**Kernel Density Estimate Plot (`kdeplot`)**

A KDE plot is used to visualize the probability density of a continuous variable.

```python
sns.kdeplot(tips['total_bill'], fill=True)
plt.title('KDE Plot of Total Bill')
plt.show()
```

### 2. Categorical Plots

**Bar Plot (`barplot`)**

Bar plots are useful for showing the relationship between a categorical variable and a continuous variable.

```python
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Average Total Bill by Day')
plt.show()
```

**Box Plot (`boxplot`)**

Box plots show the distribution of a continuous variable across different categories. It displays the median, quartiles, and potential outliers.

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Total Bill Distribution by Day')
plt.show()
```

**Violin Plot (`violinplot`)**

Violin plots are similar to box plots but also show the kernel density estimate of the distribution.

```python
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Total Bill Distribution by Day (Violin Plot)')
plt.show()
```

### 3. Relational Plots

**Scatter Plot (`scatterplot`)**

Scatter plots are used to show the relationship between two continuous variables.

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='smoker')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

**Line Plot (`lineplot`)**

Line plots are used to show the relationship between two continuous variables, with an emphasis on showing how one variable changes over time.

```python
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title('Line Plot of Total Bill vs Size')
plt.show()
```

### 4. Regression Plots

**Regression Plot (`regplot`)**

Regression plots are used to visualize the relationship between two variables and draw a linear regression model fit.

```python
sns.regplot(x='total_bill', y='tip', data=tips)
plt.title('Regression Plot of Total Bill vs Tip')
plt.show()
```

### 5. Matrix Plots

**Heatmap (`heatmap`)**

Heatmaps are used to show the correlation matrix between different variables in a dataset.

```python
# sns.heatmap(tips.corr(), annot=True, cmap='coolwarm')
# Calculate correlation matrix only for numeric columns
corr_matrix = tips.select_dtypes(include='number').corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Tips Dataset')
plt.show()
```

### 6. Pairwise Plots

**Pair Plot (`pairplot`)**

Pair plots are used to visualize pairwise relationships in a dataset. It creates a matrix of scatter plots for each pair of variables.

```python
sns.pairplot(tips, hue='sex')
plt.show()
```

### 7. Facet Grids

**FacetGrid (`FacetGrid`)**

FacetGrid is a powerful tool to create multi-plot grids based on a categorical variable.

```python
g = sns.FacetGrid(tips, col="time", row="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```
